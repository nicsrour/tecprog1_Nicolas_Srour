1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
import pandas as pd

df = pd.read_csv('alunos3.csv',sep=';',decimal=',')

df['Media_provas'] = df[df.columns[3:]].mean(axis=1)
df
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
maior = df.iloc[df['Media_provas'].idxmax(), 1]
menor = df.iloc[df['Media_provas'].idxmin(), 1]

print(f'O aluno com a menor média é: {menor}')
print(f'O aluno com a maior média é: {maior}')
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
df2 = pd.read_excel('cadastro_alunos.xlsx')
df_junto = df.merge(df2, how='left', on='RA')
df_junto
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
df_sexo = df_junto.groupby('Sexo').mean().reset_index()[['Sexo', 'Media_provas']]
df_sexo
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
df_idade = df_junto[df_junto['Media_provas']>7]['Idade'].mean()
print(f'A idade media das pessoas que obtiveram Media_provas maior ou igual a 7 é: {df_idade}')
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
df_junto.groupby('Cidade').mean().reset_index()[['Cidade', 'Media_provas']].\
sort_values(by='Media_provas',ascending=False).head(1)
```

7. Leia o arquivo `aluno3.csv`, apenas as colunas RA, Prova_1, Prova_2, Prova_3, Prova_4 calcule a médias das notas da prova e salve em um arquivo excel (xlsx) as colunas RA e Média.


```python
df = pd.read_csv('alunos3.csv',sep=';',decimal=',', usecols=['RA', 'Prova_1', 'Prova_2', 'Prova_3', 'Prova_4'])
df['Media'] = df[df.columns[1:]].mean(axis=1)
df.to_excel('alunos3.xlsx', index=False, columns=['RA','Media'])
```
